## 3 Feature Engineering

In this section you can implement the feature engineering techniques for dimensionality reduction












```python
# Train-test Split
from sklearn.model_selection import train_test_split

x = df1.drop("Loan/No Loan", axis = 1)
y = df1["Loan/No Loan"]

x_tr, x_val, y_tr, y_val = train_test_split(x, y, test_size = .3, random_state = 9)
```

```python
x_tr = x_tr.astype("float32")
x_val = x_val.astype("float32")
y_tr = y_tr.astype("float32")
y_val = y_val.astype("float32")
```

```python
# Balancing Classes
from imblearn.over_sampling import SMOTE
sm = SMOTE(sampling_strategy = 1 ,k_neighbors = 15)   #Synthetic Minority Over Sampling Technique
x_tr_res, y_tr_res = sm.fit_resample(x_tr, y_tr.ravel())
```

```python
# Applying PCA
from sklearn.decomposition import PCA as RandomizedPCA

n_components = 50              # extracting 50 most important features
pca = RandomizedPCA(n_components=n_components).fit(x_tr_res)
x_train_pca = pca.transform(x_tr_res)
x_test_pca = pca.transform(x_val)
```

```python
# Scaling Data
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(x_train_pca)
x_tr = sc.transform(x_train_pca)
x_val = sc.transform(x_test_pca)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV

lr = LogisticRegression(max_iter = 400)

param_lr = {"solver" : ["newton-cg", "lbfgs", "liblinear"]}

gs_lr = GridSearchCV(lr, param_lr, cv=10)

gs_lr.fit(x_tr, y_tr_res)
print(gs_lr.score(x_tr, y_tr_res))
lr_pr = gs_lr.predict(x_val)
print("F1 Score: ", f1_score(y_val,lr_pr))
```

```python
from sklearn.neighbors import KNeighborsClassifier

#knn = KNeighborsClassifier()

#param_knn = {"n_neighbors": range(3,11),
#            "weights" : ["uniform", "distance"],
#             "algorithm" : ["auto", "ball_tree", "kd_tree", "brute"]}

gs_knn = KNeighborsClassifier(n_neighbors = 15, weights = "uniform", algorithm = "auto")
gs_knn.fit(x_tr, y_tr_res)
knn_pr = gs_knn.predict(x_val)
print(gs_knn.score(x_val, y_val))
print(f1_score(y_val, knn_pr))
```
